## The Data
The data has been scraped from Wikipedia and is made up of two CSV files covering a span of all Super Bowls up to 2024. This data does contain missing values. The most relevant columns are noted below.  
`data/tv.csv`  
|Column|Description|
|------|-----------|
|`super_bowl`|Super Bowl number (e.g. the first Super Bowl ever is Super Bowl 1 and the last Super Bowl in 2024 is Super Bowl 58)|
|`avg_us_viewers`|Average # of US viewers|
|`share_household`|Percentage of households watching TV that watched the game|
|`rating_household`|Percentage of all households with TVs that watched the game|
|`ad_cost`|Cost per ad|
---
`data/super_bowls.csv`  
|Column|Description|
|------|-----------|
|`super_bowl`|Super Bowl number (e.g. the first Super Bowl ever is Super Bowl 1 and the last Super Bowl in 2024 is Super Bowl 58)|
|`difference_pts`|Point difference for that game|

```r
# Load packages
library(tidyverse)

# Load the CSV data
tv  <-  read_csv("data/tv.csv", show_col_types=FALSE)
super_bowls  <-  read_csv("data/super_bowls.csv", show_col_types=FALSE)

## Analyzing point differences and viewwership trends
# Create a histogram of point differences:
ggplot(super_bowls, aes(difference_pts)) +
	geom_histogram(binwidth = 2) +
	labs(x = "Point difference")

# Find smallest point difference:
smallest_point_difference <- super_bowls %>%
	filter(difference_pts == min(difference_pts))
smallest_point_difference

# Find largest point difference:
largest_point_difference <- super_bowls %>%
	filter(difference_pts == max(difference_pts))
largest_point_difference

# Relationship average viewers vs. point difference:
super_bowls_and_tv <- super_bowls %>%
	inner_join(tv, by = "super_bowl")

ggplot(super_bowls_and_tv, aes(x = difference_pts, y = avg_us_viewers)) +
	geom_point() +
	geom_smooth(method = "lm", se = FALSE) +
	labs( x = "Point difference", y = "Average US viewers")
score_impact <- "weak"

## Find trends between viewership, rating and ad costs
# Convert viewership data:
viewership_data <- tv %>%
	select(super_bowl, avg_us_viewers) %>%
	mutate(category = "Average viewers") %>%
	rename("value" = "avg_us_viewers") %>%
	relocate("value", .after = "category") %>%
	arrange(super_bowl)
viewership_data

# Convert ratings data:
ratings_data <- tv %>%
	select(super_bowl, rating_household) %>%
	mutate(category = "Household rating") %>%
	rename("value" = "rating_household") %>%
	relocate("value", .after = "category") %>%
	arrange(super_bowl)
ratings_data

# Convert advertisement data:
advertisement_data <- tv %>%
	select(super_bowl, ad_cost) %>%
	mutate(category = "Ad cost") %>%
	rename("value" = "ad_cost") %>%
	relocate("value", .after = "category") %>%
	arrange(super_bowl)
advertisement_data

# Combine data for plotting:
combined_data <- viewership_data %>%
	bind_rows(ratings_data) %>%
	bind_rows(advertisement_data)
combined_data

# Visualize trends:
ggplot(combined_data, aes(x = super_bowl, y= value, color = category)) +
	geom_line() +
	facet_wrap( ~ category) +
	labs( x = "Super Bowl")
first_to_increase <- "viewers"
```
