Analyze Nobel Prize winner data

The Nobel Prize has been among the most prestigious international awards since 1901. Each year, awards are bestowed in chemistry, literature, physics, physiology or medicine, economics, and peace. In addition to the honor, prestige, and substantial prize money, the recipient also gets a gold medal with an image of Alfred Nobel (1833 - 1896), who established the prize.

![](Nobel_Prize.png)

The Nobel Foundation has made a dataset available of all prize winners from the outset of the awards from 1901 to 2023. The dataset used in this project is from the Nobel Prize API and is available in the `nobel.csv` file in the `data` folder.

In this project, you'll get a chance to explore and answer several questions related to this prizewinning data. And we encourage you then to explore further questions that you're interested in!# Python Code

```r
# Loading in required libraries
library(tidyverse)
library(readr)
library(ggplot2)

# Start coding here!
nobel <- read_csv("data/nobel.csv")
glimpse(nobel)

# Most commonly awarded gender
most_awarded_gender <- nobel %>%
	count(sex, sort = TRUE) %>%
	arrange(desc(n))
most_awarded_gender

top_gender <- "Male"

# Most commonly awarded country of birth 
most_awarded_country <- nobel %>%
	count(birth_country, sort = TRUE) %>%
	arrange(desc(n))
most_awarded_country

top_country <- "United States of America"

# Decade with the highest proportion of US-born winners + plot
highest_prop_us_winners <- nobel %>%
	mutate(us_born = birth_country == "United States of America") %>%
	mutate(decade = floor(year/10) * 10) %>%
	group_by(decade) %>%
	summarize(proportion = mean(us_born == TRUE, na.rm = TRUE)) %>%
	arrange(desc(proportion))
highest_prop_us_winners

max_decade_usa <- "2000"

max_decade_usa_viz <- ggplot(highest_prop_us_winners, aes(decade, proportion)) +
	geom_line() +
	geom_point()
max_decade_usa_viz

# Decade and category with the highest female proportion
highest_prop_female_decade_category <- nobel %>%
	mutate(female = sex == "Female") %>%
	mutate(decade = floor(year/10) * 10) %>%
	group_by(decade, category) %>%
	summarize(proportion = mean(female == TRUE, na.rm = TRUE)) %>%
	arrange(desc(proportion))
highest_prop_female_decade_category

max_female_list <- list(decade = "2020", category = "Literature")
max_female_list

# Visualize proportion female winners
female_proportion_viz <- ggplot(highest_prop_female_decade_category, aes(decade, proportion, color = category)) +
	geom_line() +
	geom_point()
female_proportion_viz

# First woman to win a Nobel Prize and Category and list
first_woman <- nobel %>%
	filter(sex == "Female") %>%
	slice_min(year) 
first_woman

first_woman_name <- first_woman$full_name
first_woman_category <- first_woman$category

# Repeat winners and repeats list. 
repeats <- nobel %>%
	group_by(full_name) %>%
	count() %>%
	filter(n >= 2)
```
